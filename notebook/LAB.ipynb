## üß† <a name="ex1"></a>Exercici 1 ‚Äì Local vs Global

### Enunciat:
Analitza la seg√ºent funci√≥ amb m√∫ltiples m√≠nims locals i un m√≠nim global. Has de:
- Trobar anal√≠ticament el m√≠nim global.
- Escriure i programar el gradient de la funci√≥.
- Aplicar **gradient descent** des de diferents punts inicials.
- Analitzar els resultats.

### Funci√≥:

\[
f(x, y) = x^2 + y^2 + 25(\sin^2(x) + \sin^2(y))
\]

√âs una funci√≥ amb relleu ondulat, on el terme sinusoidal crea molts m√≠nims locals.

---


# üßÆ <a name="ex1"></a>Codi ‚Äì Implementaci√≥ Exercici 1

````python
import numpy as np
import matplotlib.pyplot as plt

# Definim la funci√≥
def f(x, y):
    return x**2 + y**2 + 25*(np.sin(x)**2 + np.sin(y)**2)

# Gradient de la funci√≥
def grad_f(x, y):
    df_dx = 2*x + 25*2*np.sin(x)*np.cos(x)
    df_dy = 2*y + 25*2*np.sin(y)*np.cos(y)
    return np.array([df_dx, df_dy])

# Gradient descent
def gradient_descent(starting_point, lr=0.01, steps=100):
    point = np.array(starting_point, dtype=float)
    history = [point.copy()]
    for _ in range(steps):
        grad = grad_f(point[0], point[1])
        point -= lr * grad
        history.append(point.copy())
    return np.array(history)

# Provar diferents punts inicials
starting_points = [[3, 3], [-3, -3], [5, -5], [0.1, 0.1]]
paths = [gradient_descent(p, lr=0.05, steps=100) for p in starting_points]

# Visualitzaci√≥
x = np.linspace(-6, 6, 400)
y = np.linspace(-6, 6, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.figure(figsize=(10, 6))
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
for path in paths:
    plt.plot(path[:, 0], path[:, 1], 'w.-')
plt.title('Gradient Descent paths sobre $f(x, y)$')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(label='f(x, y)')
plt.grid(True)
plt.show()
````

- La funci√≥ t√© un m√≠nim global a (x, y) ‚âà (0, 0) ja que els termes sinusoidals s√≥n m√†xims quan x o y s√≥n m√∫ltiples de œÄ.
- El gradient descent convergeix correctament si partim prop del m√≠nim global.
- Si partim lluny (ex: [5, -5]), el descens pot quedar atrapat en un m√≠nim local.
- L'aprenentatge mostra com els m√®todes basats en gradient poden tenir dificultats en funcions no convexes.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Llegim el senyal (sense cap√ßalera)
y_data = pd.read_csv('../data/signal2.csv', header=None).iloc[:, 0].values

# Fem servir x com a vector equiespaiat
x_data = np.linspace(0, 2 * np.pi, len(y_data))

# Visualitzem el senyal original
plt.figure(figsize=(10, 4))
plt.plot(x_data, y_data, label='Senyal original')
plt.title('Senyal de la gal√†xia')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

````

# üîç Pas 2: An√†lisi de Freq√º√®ncia amb FFT

````python
from scipy.fft import fft, fftfreq

# Nombre de punts
N = len(x_data)
T = (x_data[-1] - x_data[0]) / N  # Interval de mostreig

# Aplicar FFT
yf = fft(y_data)
xf = fftfreq(N, T)[:N//2]  # nom√©s part positiva

# M√≤dul de l'espectre
amplitude = 2.0/N * np.abs(yf[0:N//2])

# Visualitzaci√≥ de l'espectre
plt.figure(figsize=(10, 4))
plt.plot(xf, amplitude)
plt.title('Espectre de freq√º√®ncies (FFT)')
plt.xlabel('Freq√º√®ncia')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
````

# üìä Interpretaci√≥:
 - El gr√†fic de la FFT et mostra quines freq√º√®ncies s√≥n dominants al senyal.

 - Si veus pics n√≠tids, aix√≤ vol dir que el senyal √©s molt peri√≤dic i es pot ajustar b√© amb una o m√©s sinusoides (com sin(kx), cos(kx)).

 - Si hi ha moltes freq√º√®ncies petites, potser el senyal cont√© soroll o oscil¬∑lacions complexes.

# üßÆ Definici√≥ del model i la p√®rdua

````python
# Model amb dues sinusoides i un despla√ßament
def multi_sine_model(x, params):
    a1, b1, c1, a2, b2, c2, d = params
    return a1 * np.sin(b1 * x + c1) + a2 * np.sin(b2 * x + c2) + d

# Funci√≥ de p√®rdua (error quadr√†tic mitj√†)
def multi_sine_loss(params, x, y):
    return np.mean((multi_sine_model(x, params) - y)**2)
````

# üîÅ Gradient Descent Num√®ric (amb scipy.optimize)

````python
from scipy.optimize import minimize

# Par√†metres inicials: [a1, b1, c1, a2, b2, c2, d]
initial_params = [1, 1, 0, 0.5, 2, 0, 0]

# Optimitzaci√≥ amb 'L-BFGS-B'
result = minimize(multi_sine_loss, initial_params, args=(x_data, y_data), method='L-BFGS-B')

# Par√†metres √≤ptims trobats
opt_params = result.x
print("Par√†metres √≤ptims:", opt_params)
````

# üìà Comparaci√≥ gr√†fica

````python
# Valors ajustats
y_fit = multi_sine_model(x_data, opt_params)

# Gr√†fica comparativa
plt.figure(figsize=(10, 4))
plt.plot(x_data, y_data, label='Senyal original')
plt.plot(x_data, y_fit, label='Model amb 2 sinusoides', linestyle='--')
plt.title('Ajust del senyal amb m√∫ltiples sinusoides')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
````

# ‚úÖ Conclusions
---
- El model amb dues sinusoides pot capturar millor la forma del senyal que una sola.
- Si encara no encaixa del tot, podem afegir una tercera sinusoide.
- Aquest tipus de models permeten evitar l‚Äô√∫s d‚Äôalgoritmes m√©s costosos si l‚Äôestructura √©s clarament peri√≤dica.

# üßÆ Definici√≥ de la funci√≥ a ajustar

````python
# Proposem una funci√≥ param√®trica
def model(x, params):
    a, b, c, d = params
    return a * np.sin(b * x + c) + d

# Funci√≥ d'error (MSE)
def loss(params, x, y):
    return np.mean((model(x, params) - y) ** 2)

# Gradient descendent manual
def gradient(params, x, y, lr=0.001, steps=1000):
    hist = []
    p = np.array(params, dtype=float)
    n = len(x)
    for _ in range(steps):
        y_pred = model(x, p)
        error = y_pred - y
        
        # Derivades parcials respecte als par√†metres
        grad = np.zeros_like(p)
        grad[0] = 2/n * np.sum(error * np.sin(p[1]*x + p[2]))
        grad[1] = 2/n * np.sum(error * p[0]*x*np.cos(p[1]*x + p[2]))
        grad[2] = 2/n * np.sum(error * p[0]*np.cos(p[1]*x + p[2]))
        grad[3] = 2/n * np.sum(error)

        p -= lr * grad
        hist.append(p.copy())
    return p, hist
````

# ‚ñ∂Ô∏è Execuci√≥ i visualitzaci√≥

````python
# Par√†metres inicials: a, b, c, d
initial_params = [1.0, 1.0, 0.0, 0.0]

opt_params, history = gradient(initial_params, x_data, y_data, lr=0.001, steps=3000)

# Resultats
print("Par√†metres √≤ptims:", opt_params)

# Gr√†fica de comparaci√≥
plt.figure(figsize=(10, 4))
plt.plot(x_data, y_data, label='Senyal original')
plt.plot(x_data, model(x_data, opt_params), label='Model ajustat', linestyle='--')
plt.title('Ajust del senyal amb gradient descent')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
````

# üìä An√†lisi
---
- El model sinusoidal ajusta prou b√© el senyal.
- La funci√≥ t√© forma de $a \cdot \sin(bx + c) + d$, i hem obtingut els par√†metres √≤ptims mitjan√ßant gradient descent.
- Si el model no s‚Äôajusta prou b√©, es poden afegir m√©s termes (com una combinaci√≥ de sinusoides) o usar xarxes neuronals.

# üèπ <a name="ex4"></a>Exercici 4 ‚Äì Parashooting (Gradient Descent)

### Enunciat:
L‚Äôobjectiu √©s fer arribar una pilota a exactament 100 metres. 
La f√≥rmula de la dist√†ncia d‚Äôun tir parab√≤lic √©s:

\[
d = \frac{v_0^2 \cdot \sin(2Œ∏)}{g}
\]

Has de trobar els valors √≤ptims de:
- `v‚ÇÄ`: velocitat inicial
- `Œ∏`: angle de llan√ßament (en radians)

Utilitza **gradient descent** per trobar aquesta combinaci√≥.


# üßÆ Par√†metres i funci√≥ objectiu

````python
import numpy as np
import matplotlib.pyplot as plt

# Constants
g = 9.81  # acceleraci√≥ gravitat√≤ria
target_distance = 100  # objectiu: arribar a 100 metres

# Funci√≥ de dist√†ncia
def distance(v0, theta):
    return (v0 ** 2) * np.sin(2 * theta) / g

# Funci√≥ de p√®rdua: (dist√†ncia - 100)^2
def loss(v0, theta):
    d = distance(v0, theta)
    return (d - target_distance) ** 2
````

# üîÅ Gradient Descent

````python
# Derivades parcials
def gradients(v0, theta):
    # Partial deriv w.r.t. v0
    d_dv0 = 2 * v0 * np.sin(2 * theta) / g
    # Partial deriv w.r.t. theta
    d_dtheta = (2 * v0**2 * np.cos(2 * theta)) / g
    # Chain rule for loss
    dL_dv0 = 2 * (distance(v0, theta) - target_distance) * d_dv0
    dL_dtheta = 2 * (distance(v0, theta) - target_distance) * d_dtheta
    return dL_dv0, dL_dtheta

# Algoritme
def gradient_descent_v0_theta(v0_init, theta_init, lr=0.01, steps=1000):
    v0, theta = v0_init, theta_init
    history = []
    for _ in range(steps):
        grad_v0, grad_theta = gradients(v0, theta)
        v0 -= lr * grad_v0
        theta -= lr * grad_theta
        history.append((v0, theta, loss(v0, theta)))
    return v0, theta, history
````

# ‚ñ∂Ô∏è Execuci√≥ i visualitzaci√≥

````python
v0_init = 20
theta_init = np.pi / 4  # 45 graus

v0_opt, theta_opt, hist = gradient_descent_v0_theta(v0_init, theta_init, lr=0.001, steps=5000)

print(f"v‚ÇÄ √≤ptim = {v0_opt:.4f} m/s")
print(f"Œ∏ √≤ptim = {np.degrees(theta_opt):.2f} graus")
print(f"Dist√†ncia assolida = {distance(v0_opt, theta_opt):.4f} m")

# P√®rdua al llarg del temps
loss_values = [h[2] for h in hist]
plt.figure(figsize=(8, 4))
plt.plot(loss_values)
plt.title("Evoluci√≥ de l'error durant el gradient descent")
plt.xlabel("Iteraci√≥")
plt.ylabel("Error quadr√†tic")
plt.grid(True)
plt.show()
````

# ‚úÖ Conclusions
---
- El gradient descent ha trobat una combinaci√≥ de velocitat i angle que dona una dist√†ncia molt propera als 100 m.
- Podem modificar `lr` (learning rate) i `steps` per ajustar el rendiment.
- Notar que Œ∏ ha de mantenir-se dins del rang f√≠sic [0, œÄ/2] (0 a 90¬∞).

# üîß Implementaci√≥ amb DEAP
### 1. Configuraci√≥ Inicial

````python
import random
from deap import base, creator, tools
import numpy as np
````

````python
# ================= CONFIGURACI√ì INICIAL =================
packages = [25, 46, 25, 11, 34, 48, 85, 120, 111, 70, 87, 35, 61, 102, 94, 131, 73, 9, 142]
max_capacity = 240
````

````python
# ================= ALGORITME PRINCIPAL =================
def find_min_containers(packages, max_capacity):
    # Ordenem els paquets de major a menor
    sorted_packages = sorted(packages, reverse=True)
    
    # Funci√≥ per comprovar si una assignaci√≥ √©s v√†lida
    def is_valid_assignment(num_containers, assignment):
        containers = [0] * num_containers
        for pkg, cont in zip(sorted_packages, assignment):
            containers[cont] += pkg
            if containers[cont] > max_capacity:
                return False
        return True
    
    # Cerca bin√†ria per trobar el m√≠nim nombre de contenidors
    left = 1
    right = len(sorted_packages)
    best_solution = None
    
    while left <= right:
        mid = (left + right) // 2
        found = False
        
        # Prova m√∫ltiples assignacions per aquest nombre de contenidors
        for _ in range(10):  # N√∫mero d'intents per cada mida
            # Genera una assignaci√≥ aleat√≤ria
            assignment = [random.randint(0, mid-1) for _ in sorted_packages]
            
            # Optimitza l'assignaci√≥ amb algorisme gen√®tic
            optimized_assignment = optimize_assignment(mid, sorted_packages, assignment.copy(), max_capacity)
            
            if is_valid_assignment(mid, optimized_assignment):
                best_solution = (mid, optimized_assignment)
                found = True
                break
        
        if found:
            right = mid - 1  # Prova amb menys contenidors
        else:
            left = mid + 1   # Necessitem m√©s contenidors
    
    return best_solution
````

### 2. OPTIMITZACI√ì PER ALGORISME GEN√àTIC

````python
# ================= OPTIMITZACI√ì PER ALGORISME GEN√àTIC =================
def optimize_assignment(n_containers, packages, initial_assignment, max_capacity):
    # Configuraci√≥ DEAP
    creator.create("FitnessMax", base.Fitness, weights=(1.0,))
    creator.create("Individual", list, fitness=creator.FitnessMax)
    
    toolbox = base.Toolbox()
    toolbox.register("attr_int", random.randint, 0, n_containers-1)
    toolbox.register("individual", tools.initIterate, creator.Individual, 
                    lambda: [random.randint(0, n_containers-1) for _ in packages])
    toolbox.register("population", tools.initRepeat, list, toolbox.individual)
    
    # Funci√≥ d'avaluaci√≥
    def evaluate(individual):
        containers = [0] * n_containers
        for pkg, cont in zip(packages, individual):
            containers[cont] += pkg
        
        # Penalitzaci√≥ per sobrec√†rrega
        overflow = sum(max(0, load - max_capacity) for load in containers)
        
        # Recompensa per utilitzaci√≥ equilibrada
        utilization = sum(min(load, max_capacity) for load in containers)
        balance = 1 - (np.std([load for load in containers if load <= max_capacity]) / max_capacity 
                      if len([load for load in containers if load <= max_capacity]) > 1 else 0)
        
        return (utilization * balance - 1000 * overflow,)
    
    toolbox.register("evaluate", evaluate)
    toolbox.register("mate", tools.cxTwoPoint)
    toolbox.register("mutate", tools.mutUniformInt, low=0, up=n_containers-1, indpb=0.1)
    toolbox.register("select", tools.selTournament, tournsize=3)
    
    # Execuci√≥ de l'algorisme gen√®tic
    population = toolbox.population(n=100)
    if initial_assignment:
        population[0] = creator.Individual(initial_assignment)
    
    algorithms.eaSimple(population, toolbox, cxpb=0.7, mutpb=0.2, ngen=50, verbose=False)
    
    return tools.selBest(population, k=1)[0]
````

# üìä Visualitzaci√≥ dels Resultats

````python
# ================= EXECUCI√ì I VISUALITZACI√ì =================
def print_solution(num_containers, assignment, packages):
    containers = [[] for _ in range(num_containers)]
    for pkg, cont in zip(packages, assignment):
        containers[cont].append(pkg)
    
    print(f"\nüîç SOLUCI√ì √íPTIMA: {num_containers} contenidors necessaris")
    print("="*40)
    total_valid = 0
    for i, cont in enumerate(containers):
        load = sum(cont)
        status = "‚úÖ V√ÄLID" if load <= max_capacity else "‚ùå EXC√âS"
        print(f"Contenidor {i+1}: {load:>3} kg ({load/max_capacity:>5.1%}) {status}")
        print(f"Paquets: {sorted(cont)}")
        if load <= max_capacity:
            total_valid += load
    
    print(f"\nüöÄ C√ÄRREGA TOTAL V√ÄLIDA: {total_valid} kg / {num_containers*max_capacity} kg")
    print(f"üìä EFICI√àNCIA: {total_valid/(num_containers*max_capacity):.1%}")
````

En aquest exercici gestionem una flota de 5 taxis i volem assignar-los als 5 clients disponibles.  
Disposem dels temps d'arribada estimats (ETA) de cada taxi a cada client en una taula.  
L'objectiu √©s **assignar un taxi a cada client** de manera que es **minimitzi el temps d'espera mitj√† dels clients**.

Farem servir l'**algorisme h√∫ngar** per trobar l'assignaci√≥ √≤ptima.

````python
import numpy as np
from scipy.optimize import linear_sum_assignment

# Matriu dels temps ETA (fila: taxis, columna: clients)
eta = np.array([
    [5, 7, 3, 5, 8],  # T1
    [3, 2, 1, 4, 6],  # T2
    [2,10, 8, 6, 7],  # T3
    [4, 6, 4, 5, 9],  # T4
    [5, 8, 2, 8, 8]   # T5
])

# Algorisme h√∫ngar per minimitzar la suma de temps
row_ind, col_ind = linear_sum_assignment(eta)

# Mostrar l'assignaci√≥ √≤ptima i el temps total i mitj√†
print("üü¢ Assignaci√≥ √≤ptima Taxi ‚Üí Client:")
for taxi, client in zip(row_ind, col_ind):
    print(f"  üöï T{taxi+1} ‚Üí C{client+1} (ETA: {eta[taxi, client]} min)")

temps_total = eta[row_ind, col_ind].sum()
temps_mig = temps_total / len(row_ind)

print(f"\nüïí Temps total d‚Äôespera: {temps_total} min")
print(f"‚è≥ Temps mitj√† d‚Äôespera: {temps_mig:.2f} min")
````

### 3.1 Explicaci√≥ te√≤rica
Els algorismes gen√®tics simulen l'evoluci√≥ natural:
1. Poblaci√≥ inicial de solucions (cromosomes)
2. Selecci√≥ dels millors (fitness)
3. Creuament (crossover)
4. Mutaci√≥
5. Nova generaci√≥

````python
# Cel¬∑la de codi amb la implementaci√≥ GA
def genetic_algorithm_tsp(cities, pop_size=100, elite_size=20, mutation_rate=0.01, generations=500):
    """Implementaci√≥ de GA per al TSP"""
    population = [generate_random_route(cities) for _ in range(pop_size)]
    
    for gen in range(generations):
        population = next_generation(population, elite_size, mutation_rate)
    
    best_route = min(population, key=calculate_distance)
    return best_route, calculate_distance(best_route)

def next_generation(population, elite_size, mutation_rate):
    """Genera la seg√ºent generaci√≥ de solucions"""
    ranked = sorted(population, key=calculate_distance)
    elites = ranked[:elite_size]
    
    children = []
    for _ in range(len(population) - elite_size):
        parent1, parent2 = selection(ranked), selection(ranked)
        child = crossover(parent1, parent2)
        child = mutate(child, mutation_rate)
        children.append(child)
    
    return elites + children
````
